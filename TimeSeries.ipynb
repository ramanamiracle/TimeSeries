```python
dd= np.asarray(train['value'])
y_hat = test.copy()
y_hat['naive'] = dd[len(dd)-1]
plt.figure(figsize=(12,8))
plt.plot(train.index, train['value'], label='Train')
plt.plot(test.index,test['value'], label='Test')
plt.plot(y_hat.index,y_hat['naive'], label='Naive Forecast')
plt.legend(loc='best')
plt.title("Naive Forecast")
```

```python
dd
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(test.value, y_hat.naive))
print(rms)
```

```python
#Naive
y_hat_avg = test.copy()
y_hat_avg['avg_forecast'] = train['value'].mean()
plt.plot(train['value'], label='Train')
plt.plot(test['value'], label='Test')
plt.plot(y_hat_avg['avg_forecast'], label='Average Forecast')
plt.legend(loc='best')
```

```python
rms = sqrt(mean_squared_error(test['value'], y_hat_avg.avg_forecast))
print(rms)
```

```python
#Movine Average
y_hat_avg = test.copy()
y_hat_avg['moving_avg_forecast'] = train['value'].rolling(2).mean()
#y_hat_avg['moving_avg_forecast'] = y_hat_avg.rolling(2).mean()
plt.figure(figsize=(16,8))
plt.plot(train['value'], label='Train')
plt.plot(test['value'], label='Test')
plt.plot(y_hat_avg['moving_avg_forecast'].dropna(), label='Moving Average Forecast')
plt.legend(loc='best')
plt.plot(train['value'].rolling(2).mean() , label='Moving Average Forecast')
plt.plot(test['value'].rolling(2).mean() , label='Moving Average Forecast')
```
